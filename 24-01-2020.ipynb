# array range using arange(array range)

```python
print(np.arange(1,10)) # dynamic array
print(np.arange(10))# starts from zero
print(np.arange(1,10,2))# alternate numbers
print(np.arange(9,0,-2))# in reverse order
print(np.arange(24).reshape(3,8))# to arrange elements into dimensions of required (rows:coloum) 
```

```python
# linspace()
np.linspace(1,10,10)
```

```python
# to print random numbers in array
#print(np.random.randint(1,10,5).reshape(5,1))
print(np.random.randint(1,10,(3,3)))
```

```python
# to print normal distribution values randomly
print(np.random.randn(3))
print(np.random.randn(3,2))
print(np.random.random((3,2)))
```

```python
# array initialisation
#'''- np.ones()
#- np.zeros()
#- np.diag()
#- np.full()'''

o=np.ones((2,3,4))
u=np.zeros((2,3,4))
e=np.diag((2,3,4))
r=np.full((3,3),4,)
print(o)
print(u)
print(r)
```
